```python
from sklearn.feature_extraction import image
import image
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

```

```python
I = plt.imread('subject1_bottom.tiff')
plt.imshow(I,aspect="auto")
plt.show()
```

```python
# Use the array data from the first image in this dataset
one_image = I; 
print(one_image)
print('Image shape: {}'.format(one_image.shape))

```

```python
from sklearn.feature_extraction import image
patches = image.extract_patches_2d(one_image, (64, 64)) # here I am Extracting patches of 64*64
print('Patches shape: {}'.format(patches.shape))


```

```python
# from the above shape you can see we got 201601 patches from 1 image of size 64*64
print(patches[100000])
```

```python
# print some patches
%matplotlib inline 

for i in range(100000,100100):
    plt.imshow(patches[i],aspect='auto' )
    plt.show()


```
